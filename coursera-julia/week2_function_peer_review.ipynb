# A function of one variable:

- [example](https://www.wolframalpha.com/input/?i=plot+x%5E3+-+6x%5E2+%2B+4x+%2B+12&lk=3)

```julia
function one_variable(x)
    y = x^3 - 6x^2 + 4x + 12
    return y
end
```

```julia
one_variable(10)
```

- The function looks like a sigmoid function

$$ f(x) = x^3 - 6x^2 + 4x + 12 $$

```julia
start = 1
num = -36:36
total = length(num)
result = Array{Union{Nothing, Int}}(nothing, total, 2)
for num = -36:36
#     println(start)
    result[start,1] = num
    result[start,2] = one_variable(num)
    println(num, ", ", one_variable(num))
    start = start + 1
end
```

```julia
using Plots
gr()

plot(result[:, 1], result[:, 2])
```
